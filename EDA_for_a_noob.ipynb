```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
##Data types info
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##Pie Chart
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Maximum transactions are from India followed by USA then United Kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. When rating is between 4.5 to 4.9----> Excellent
1. When rating is between 4.0 to 4.4----> Very Good
1. When rating is between 3.5 to 3.9----> Good
1. When rating is between 2.5 to 3.4----> Average
1. When rating is between 0.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x='Aggregate rating',hue= 'Aggregate rating', y= 'Rating Count',data=ratings,palette='mako',legend=False)
```

```python
sns.barplot(x='Aggregate rating', y= 'Rating Count',hue= 'Rating Count',data=ratings,palette='dark',legend=False) 
```

```python
## Count plot
sns.countplot(x='Rating color',data=ratings, palette='dark',hue='Rating color')
```

```python
##Find the countries that have given 0 ratings
final_df.groupby(['Aggregate rating','Country']).size().reset_index().rename(columns={0:'Count'}).head()
```

```python
##Find out which currency is used by which country?
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
##Which countries do have online delivery option?
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```
